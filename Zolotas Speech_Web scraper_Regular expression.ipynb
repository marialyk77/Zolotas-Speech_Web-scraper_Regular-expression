```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://de.wikipedia.org/wiki/Xenophon_Zolotas'

page = requests.get(url)

soup = BeautifulSoup(page.text, 'html')
```

```python
print(soup)
```

```python
Jolotas_speech = soup.find_all('p')

Jolotas_speech
```

```python
# All paragraphs have the label 'p', but i am interested to count the number of word's occurences only for the past paragraph

# So, i have to iterate through all p to find the one i want 
```

```python
target_paragraph_text = None ## In this variable I am going to save the paragraph i want to pull.

for paragraph in soup.find_all('p'):
    if "Kyrie, it is Zeus’ anathema on our epoch" in paragraph.text: # checking to see if the paragraph contains the desired text
        target_paragraph_text = paragraph.text
        break
        
print(target_paragraph_text)       
```

```python
type(target_paragraph_text)
```

```python
# For removing punctuation, i will use Regular Expression
```

```python
import re 
```

```python
speech_no_punt = re.sub(r'[^\w\s]', '',target_paragraph_text) # to remove punct and keep letters and white space

speech_no_punt
```

```python
## I still see some issues: A. ' - single quote , B. \n - new line character 
```

```python
speech_cleaned = speech_no_punt.strip() ## to remove the \n

if speech_cleaned.startswith("'"): ## string slicing to remove the : '
    speech_cleaned = speech_cleaned[1:]

zolotas_speech_lower = speech_cleaned.lower()

if zolotas_speech_lower.endswith("'"): ## to remove the : ' at the end
    zolotas_speech_lower = zolotas_speech_lower[:-1]


print(zolotas_speech_lower)

```

```python
# for seperating every word in its own value , before putting in a df 
```

```python
re.split(r'\s+', zolotas_speech_lower)
```

```python
zolotas_speech_broken_out = re.split(r'\s+', zolotas_speech_lower)

zolotas_speech_broken_out
```

```python
import pandas as pd 
```

```python
df = pd.DataFrame(zolotas_speech_broken_out).value_counts()

df
```
